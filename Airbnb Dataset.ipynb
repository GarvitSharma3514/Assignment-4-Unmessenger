```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("C:\\Users\\garvit\\Desktop\\Airbnb_Dataset.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df['price'] = df['price'].astype(int)
df['minimum_nights'] = df['minimum_nights'].astype(int)
df['number_of_reviews'] = df['number_of_reviews'].astype(int)
df['calculated_host_listings_count'] = df['calculated_host_listings_count'].astype(int)
df['availability_365'] = df['availability_365'].astype(int)
```

```python
df.info()
```

# 1. Distribution of Listings by Neighbourhood Group

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='neighbourhood_group', palette='viridis')
plt.title('Distribution of Listings by Neighbourhood Group')
plt.ylabel('Number of Listings')
plt.xlabel('Neighbourhood Group')
plt.xticks(rotation=45)
plt.show()
```

# 2. Price Distribution by Room Type

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='room_type', y='price', palette='viridis')
plt.title('Price Distribution by Room Type')
plt.ylabel('Price')
plt.xlabel('Room Type')
plt.yscale('log') 
plt.show()
```

# 3. Number of Reviews per Month

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='reviews_per_month', bins=30, kde=True, color='skyblue')
plt.title('Number of Reviews per Month')
plt.ylabel('Frequency')
plt.xlabel('Reviews per Month')
plt.show()
```

# 5. Availability Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='availability_365', bins=30, kde=True, color='salmon')
plt.title('Availability Distribution')
plt.ylabel('Frequency')
plt.xlabel('Availability (days per year)')
plt.show()
```

# 5. Average Price per Neighbourhood Group

```python
plt.figure(figsize=(10, 6))
avg_price_neighbourhood = df.groupby('neighbourhood_group')['price'].mean().reset_index()
sns.barplot(data=avg_price_neighbourhood, x='neighbourhood_group', y='price', palette='viridis')
plt.title('Average Price per Neighbourhood Group')
plt.ylabel('Average Price')
plt.xlabel('Neighbourhood Group')
plt.xticks(rotation=45)
plt.show()
```

# Remarks on Visualizations


Distribution of Listings by Neighbourhood Group:

Column Used: neighbourhood_group
Purpose: To show the distribution of Airbnb listings across different neighbourhood groups. This helps identify which areas have the most listings.
Price Distribution by Room Type:

Columns Used: room_type, price
Purpose: To compare the price ranges for different types of rooms. A box plot is used to visualize the spread and identify any outliers.
Average Price per Neighbourhood Group:

Columns Used: neighbourhood_group, price
Purpose: To show the average price of listings in each neighbourhood group. This bar plot highlights which areas are more expensive on average.
Number of Reviews per Month:

Column Used: reviews_per_month
Purpose: To visualize the distribution of the number of reviews per month across listings. This helps understand the review frequency.
Availability Distribution:

Column Used: availability_365
Purpose: To show how many days in a year the listings are available. This histogram helps identify common availability patterns among listings.

    
These visualizations provide insights into the distribution, pricing, and availability of Airbnb listings, helping hosts and guests make informed decisions.
