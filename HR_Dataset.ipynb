# 1. Distribution of Employees by Department

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Department', palette='viridis')
plt.title('Distribution of Employees by Department')
plt.ylabel('Number of Employees')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.show()
```

# 2. Salary Distribution by Gender

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Sex', y='Salary', palette='viridis')
plt.title('Salary Distribution by Gender')
plt.ylabel('Salary')
plt.xlabel('Gender')
plt.show()
```

# 3. Average Salary by Department

```python
plt.figure(figsize=(10, 6))
avg_salary_department = df.groupby('Department')['Salary'].mean().reset_index()
sns.barplot(data=avg_salary_department, x='Department', y='Salary', palette='viridis')
plt.title('Average Salary by Department')
plt.ylabel('Average Salary')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.show()
```

# 4. Performance Scores by Department

```python

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Department', y='PerfScoreID', palette='viridis')
plt.title('Performance Scores by Department')
plt.ylabel('Performance Score')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.show()
```

# 5. Number of Absences

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Absences', bins=30, kde=True, color='salmon')
plt.title('Number of Absences')
plt.ylabel('Frequency')
plt.xlabel('Absences')
plt.show()
```
